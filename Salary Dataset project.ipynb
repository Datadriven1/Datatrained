```python
from sklearn.impute import SimpleImputer
```

```python
imp=SimpleImputer(strategy='most_frequent')
sl['City']=imp.fit_transform(sl['City'].values.reshape(-1,1))
sl
```

```python
imp1=SimpleImputer(strategy='mean')
sl['Age']=imp1.fit_transform(sl['Age'].values.reshape(-1,1))
sl['Salary']=imp1.fit_transform(sl['Salary'].values.reshape(-1,1))
```

```python
sl
```

```python
sl.isnull().sum()
```

```python
sl.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
list1=['City','Country']

for val in list1:
    sl[val]=le.fit_transform(sl[val].astype(str))
```

```python
x=sl.drop(columns=['Salary', 'Name','SNo'])
y=sl[['Salary']]
```

```python
x
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
lm=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.22)
```

```python
lm.fit(x_train,y_train)
```

```python
lm.score(x_train,y_train)
```

```python
pred=lm.predict(x_test)
pred
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from sklearn.metrics import r2_score
print(r2_score(y_test,pred))
```

```python
print('error:')
print('mean absolute error:', mean_absolute_error(y_test,pred))
print('mean squared error:', mean_squared_error(y_test,pred))

print('root mean squared error:', np.sqrt(mean_squared_error(y_test,pred)))
```

```python
import joblib
joblib.dump(lm,'lm_salary.obj')
```
